```python
proteins = []
drugs = []
diseases = []
proteins_txt = open("../data/protein.txt","r")
for protein in proteins_txt:
    proteins.append(protein.replace("\n",""))
proteins_txt.close()
drugs_txt = open("../data/drug.txt","r")
for drug in drugs_txt:
    drugs.append(drug.replace("\n",""))
drugs_txt.close()
diseases_txt = open("../data/disease.txt","r")
for disease in diseases_txt:
    diseases.append(disease.replace("\n",""))
diseases_txt.close()
drug_protein_matrix = np.loadtxt('../data/mat_drug_protein.txt')
drug_protein_matrix = drug_protein_matrix.astype('int64')
drug_similarities = np.loadtxt("../data/Similarity_Matrix_Drugs.txt")
protein_similarities = np.loadtxt("../data/Similarity_Matrix_Proteins.txt")
protein_drug_matrix = np.loadtxt('../data/mat_protein_drug.txt')
drug_drug_matrix = np.loadtxt('../data/mat_drug_drug.txt')
drug_disease_matrix=np.loadtxt('../data/mat_drug_disease.txt')
protein_disease_matrix=np.loadtxt('../data/mat_protein_disease.txt')
drug_feature_vector = np.loadtxt('../feature/drug_vector_d100.txt')
protein_feature_vector = np.loadtxt('../feature/protein_vector_d400.txt')
```

```python
drug_protein_matrix = torch.tensor(drug_protein_matrix,dtype=torch.int8)
drug_disease_matrix = torch.tensor(drug_disease_matrix,dtype=torch.int8)
protein_disease_matrix = torch.tensor(protein_disease_matrix,dtype=torch.int8)
drug_drug_matrix = torch.tensor(drug_drug_matrix,dtype=torch.float32)
protein_drug_matrix = torch.tensor(protein_drug_matrix,dtype=torch.int8)
drug_similarities = torch.tensor(drug_similarities,dtype=torch.float32)
protein_similarities = torch.tensor(protein_similarities,dtype=torch.float32)
drug_feature_vector = torch.tensor(drug_feature_vector,dtype=torch.float32)
protein_feature_vector = torch.tensor(protein_feature_vector,dtype=torch.float32)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

B = nx.Graph()
B.add_nodes_from([i for i in range(len(drugs))], bipartite=0)
B.add_nodes_from([i for i in range(len(drugs)+len(proteins))], bipartite=1)
edges = []
for i in range(len(drugs)):
    for j in range(len(proteins)):
        if drug_protein_matrix[i, j] == 1:
            edges.append((i, j + len(drugs)))
B.add_edges_from(edges)
```

```python
# Draw the bipartite graph
plt.figure(figsize=(15, 10))
pos = nx.spring_layout(B, k=0.15)
nx.draw(B, pos, with_labels=True, node_color=['skyblue' if node in range(len(drugs)) else 'salmon' for node in B.nodes()], node_size=50, edge_color='gray', font_size=8)
plt.title("Bipartite Graph of Drug-Protein Interactions")
plt.show()
```

```python
from node2vec import Node2Vec
# Precompute probabilities and generate walks
node2vec = Node2Vec(B, dimensions=64, walk_length=30, num_walks=200, workers=4)
# Embed nodes
model = node2vec.fit(window=10, min_count=1, batch_words=4)

# Get embeddings for nodes
node_embeddings = model.wv
print(node_embeddings)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATConv
from sklearn.metrics import roc_auc_score


num_drugs = drug_feature_vector.shape[0]
num_proteins = protein_feature_vector.shape[0]
class GNNPredictor(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_heads, dropout=0.5):
        super(GNNPredictor, self).__init__()
        self.gnn = GATConv(input_dim, hidden_dim, heads=num_heads, concat=False, dropout=dropout)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, edge_index, num_drugs, num_proteins):
        x = self.gnn(x, edge_index)
        x = self.dropout(x)
        interaction_logits = torch.matmul(x[:num_drugs], x[num_drugs:num_drugs + num_proteins].T)
        interaction_probs = torch.sigmoid(interaction_logits)
        return interaction_probs

# Example usage
# Convert node embeddings to a tensor
all_nodes = list(B.nodes())
embedding_matrix = np.array([node_embeddings[str(node)] for node in all_nodes])
embedding_tensor = torch.tensor(embedding_matrix, dtype=torch.float32)
# Create edge_index tensor
edge_index = torch.tensor([[all_nodes.index(edge[0]), all_nodes.index(edge[1])] for edge in B.edges()], dtype=torch.long).t().contiguous()
# Initialize and train the model
input_dim = embedding_tensor.shape[1]
hidden_dim = 128
output_dim = 64
num_heads = 4
dropout = 0.5

model = GNNPredictor(input_dim, hidden_dim, num_heads, dropout)
optimizer = optim.Adam(model.parameters(), lr=0.0005)
criterion = nn.BCELoss()
model.train()
for epoch in range(100):
    total_loss = 0
    all_preds = []
    all_labels = []
    optimizer.zero_grad()
    prediction = model(embedding_tensor, edge_index, num_drugs, num_proteins).flatten()
    label = drug_protein_matrix.type(torch.float32).flatten()
    #for i in range(num_drugs):
    #    for j in range(num_proteins):
    #        label = drug_protein_matrix[i][j]
    #        label = label.type(torch.float32)
    #        prediction = output[i, j+len(drugs)]
    print(f'prediction = {prediction.shape}')
    print(f'label = {label.shape}')
    loss = criterion(prediction, label)
    print(f'loss = {loss}')
    loss.backward()
    optimizer.step()
    print(prediction.detach().numpy())
    #auroc = roc_auc_score(prediction.detach().numpy(), label)
    print(f'Epoch {epoch+1}, Loss: {loss}, AUROC: auroc:.4')
```
